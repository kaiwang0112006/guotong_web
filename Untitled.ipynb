```python
import pandas as pd
```

```python
df = pd.read_excel("processed_knowledge_base_20250919_155852.xlsx")
df.head()
```

```python
result = []
for row in df.itertuples():
    #print(row.name)
    #print(row.description)
    deslist = row.description.split("\n")
    itemresult = []
    for itstr in deslist:
        if "—" in itstr:
            itstr = itstr.replace("—","")
            if "：" in itstr:
                items = itstr.split("：")[-1].split("、")
            else:
                if "其他" not in itstr:
                    items = [itstr.replace("；","")]
            itemresult += items
    for i in itemresult:
        i = i.replace("；","")
        result.append({"name":row.name, "product":i})
```

```python
resultdf = pd.DataFrame(result)
resultdf.head()
```

```python
resultdf.to_csv("processed_knowledge_peoduct.csv",index=False)
```
